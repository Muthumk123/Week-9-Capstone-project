```python
model = pickle.load(open("svm_model_final_capstone_project.pkl", "rb"))
scaler = pickle.load(open("scaler.pkl", "rb"))
columns = pickle.load(open("columns.pkl", "rb"))
user_input = {
    'age': 52,
    'sex': 1,
    'cp': 2,
    'chol': 230,
    'fbs': 0,
    'restecg': 1,
    'thalach': 150,
    'exang': 0,
    'oldpeak': 1.2,
    'slope': 1,
    'ca': 0,
    'thal': 2
}
user_df = pd.DataFrame([user_input])
user_df = pd.get_dummies(user_df) # create dummy columns
user_df = user_df.reindex(columns=columns, fill_value=0) # align columns
user_scaled = scaler.transform(user_df) # scale
prediction = model.predict(user_scaled)
print("Prediction:", prediction)
if prediction[0] == 1:
    print("Heart Disease Positive")
else:
    print("Heart Disease Negative")
```
